```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('WineQT.csv')

# Move the 'id' column to the front of the DataFrame
df.insert(0, 'Id', df.pop('Id'))

# Select the desired columns
df = df[['Id', 'pH', 'sulphates', 'alcohol', 'quality']]

# Filter the DataFrame to only include quality ratings of 3, 5, and 7
df = df[df['quality'].isin([3, 5, 7])]

# Sort the DataFrame by quality rating in descending order
df = df.sort_values(by='quality', ascending=False)

# Randomly sample 5 rows from the DataFrame
sampled_df = df[['Id']].sample(n=25)

# Display the resulting DataFrame
sampled_df

import pandas as pd

# Read the CSV file
df = pd.read_csv('WineQT.csv')

# Move the 'id' column to the front of the DataFrame
df.insert(0, 'Id', df.pop('Id'))

# Select the desired columns
df = df[['Id', 'pH', 'sulphates', 'alcohol', 'quality']]

# Filter the DataFrame to only include quality ratings of 3, 5, and 7
df = df[df['quality'].isin([3, 5, 7])]

# Sort the DataFrame by quality rating in descending order
df = df.sort_values(by='quality', ascending=False)

# Randomly sample 5 rows from the DataFrame
sampled_df = df[['Id']].sample(n=25)

# Display the resulting DataFrame
sampled_df





```

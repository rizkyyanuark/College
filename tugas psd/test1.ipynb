```python
import numpy as np
import scipy.signal as signal
import scipy.io.wavfile as wavfile
import matplotlib.pyplot as plt
```

```python
# Load audio file
Fs, x = wavfile.read('psd cover 2.wav')

# Create time vector
t = np.arange(len(x)) / Fs

# Design filter
f = [0, 0.2, 0.3, 1]
m = [1, 1, 0, 0]
b = signal.firwin2(100, f, m)
```

```python
plt.plot(t, x)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()

# Apply filter
y = signal.lfilter(b, 1, x)

# Save audio file
wavfile.write('audio_file_filtered.wav', Fs, y)
```

```python
# Load audio file
rate, data = wavfile.read('psd cover 2.wav')

# Apply FFT to audio data
fft = np.fft.fft(data)

# Find the frequency of the noise
noise_freq = 1 # Change this to the frequency of the noise in your audio file

# Remove the noise from the audio data
fft[:, noise_freq] = 0

# Apply inverse FFT to get cleaned audio data
clean_data = np.fft.ifft(fft)

# Save cleaned audio data to file
wavfile.write('clean_audio_file.wav', rate, clean_data)
```

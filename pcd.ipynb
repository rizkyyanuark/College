```python

# Import the necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import re
# Load the image
image = cv2.imread(r"C:\Users\rizky\OneDrive\Desktop\WhatsApp Image 2023-03-15 at 21.06.11_cb332051.jpg")

# Convert BGR image to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert RGB image to Grayscale
grayscale_img = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Convert RGB image to HSV
hsv_img = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2HSV)

# Create subplots
fig, axs = plt.subplots(1, 4, figsize=(15, 4))  # Mengubah simbol '-' menjadi '='

# Plot the original image
axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')  # Mengubah string pada title

# Plot the grayscale image
axs[1].imshow(grayscale_img, cmap='gray')
axs[1].set_title('Grayscale Image')  # Mengubah string pada title

# Plot the HSV image
axs[2].imshow(hsv_img)
axs[2].set_title('HSV Image')  # Mengubah string pada title

# Remove ticks from the subplots
for ax in axs:
    ax.set_xticks([])
    ax.set_yticks([])

# Display the subplots
plt.tight_layout()
plt.show()
```

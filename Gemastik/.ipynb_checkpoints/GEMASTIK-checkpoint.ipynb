```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import itertools
# !pip install pandas
# !pip install git+https://github.com/JustAnotherArchivist/snscrape.git
```

```python
queries = ['korupsi']
```

```python
# Jumlah data yang akan di-scrape
limit = 15000

# Tenggat waktu awal dan akhir (format: YYYY-MM-DD)
start_date = '2021-01-01'
end_date = '2023-06-27'

# Query pencarian
query_list = ['korupsi',
           'tindak korupsi',
           'kasus korupsi',
           'penyuapan',
           'suap',
           'mark up',
           'kolusi',
           'nepotisme',
           'gratifikasi',
           'dugaan korupsi',
           'koruptor',
           'KPK',
           'dana korupsi',
           'penggelapan uang',
           'penyimpangan anggaran',
           'aliran dana ilegal',
           'suap proyek',
           'mafia korupsi']

# List untuk menyimpan data
tweets = []

# Melakukan scraping
for query in query_list:
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{query} since:{start_date} until:{end_date}').get_items()):
        # Menambahkan data ke list
        tweets.append([tweet.user.username, tweet.content, tweet.date])

        # Memeriksa apakah jumlah data telah mencapai batas
        if i + 1 == limit:
            break

    # Memeriksa apakah jumlah data telah mencapai batas
    if len(tweets) >= limit:
        break
```

```python
df = pd.DataFrame(tweets, columns=['User', 'Isi konten', 'Date'])
```

```python
import pandas as pd
import re

# Load datasets
datakorupsibaru = pd.read_csv("datakorupsi.csv", encoding='latin-1')
new_kamusalay = pd.read_csv("new_kamusalay.csv", encoding='latin-1')
kamus_singkatan_deluxe = pd.read_csv("kamus_singkatan_deluxe.csv", encoding='latin-1')

# Create a dictionary of slang words from new_kamusalay.csv
slang_dict = dict(zip(new_kamusalay['anakjakartaasikasik'], new_kamusalay['anak jakarta asyik asyik']))

# Create a dictionary of abbreviations from kamus_singkatan_deluxe.csv
abbreviation_column = input("Enter the name of the column containing abbreviations: ")
full_form_column = input("Enter the name of the column containing full forms: ")
abbreviation_dict = dict(zip(kamus_singkatan_deluxe[abbreviation_column], kamus_singkatan_deluxe[full_form_column]))

# Function to clean the text
def clean_text(text):
    # Remove special characters and links
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Convert text to lowercase
    text = text.lower()

    # Remove hashtags
    text = re.sub(r'#', '', text)

    # Clean slang words
    text = ' '.join([slang_dict.get(word, word) for word in text.split()])

    # Clean abbreviations
    text = ' '.join([abbreviation_dict.get(word, word) for word in text.split()])

    return text

# Apply the clean_text function to the 'Isi konten' column in datakorupsibaru
datakorupsibaru['Isi konten'] = datakorupsibaru['Isi konten'].apply(clean_text)

# Remove duplicates
datakorupsibaru.drop_duplicates(subset=['Isi konten'], inplace=True)

# Save the cleaned data to a new CSV file
datakorupsibaru.to_csv("cleaned_datakorupsibaru.csv", index=False)
```

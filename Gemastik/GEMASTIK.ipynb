```python
# Jumlah data yang akan di-scrape
limit = 15000

# Tenggat waktu awal dan akhir (format: YYYY-MM-DD)
start_date = '2021-01-01'
end_date = '2023-06-27'

# Query pencarian
query_list = ['korupsi',
           'tindak korupsi',
           'kasus korupsi',
           'penyuapan',
           'suap',
           'mark up',
           'kolusi',
           'nepotisme',
           'gratifikasi',
           'dugaan korupsi',
           'koruptor',
           'KPK',
           'dana korupsi',
           'penggelapan uang',
           'penyimpangan anggaran',
           'aliran dana ilegal',
           'suap proyek',
           'mafia korupsi']

# List untuk menyimpan data
tweets = []

# Melakukan scraping
for query in query_list:
    for i, tweet in enumerate(sntwitter.TwitterSearchScraper(f'{query} since:{start_date} until:{end_date}').get_items()):
        # Menambahkan data ke list
        tweets.append([tweet.user.username, tweet.content, tweet.date])

        # Memeriksa apakah jumlah data telah mencapai batas
        if i + 1 == limit:
            break

    # Memeriksa apakah jumlah data telah mencapai batas
    if len(tweets) >= limit:
        break
```

```python
df = pd.DataFrame(tweets, columns=['user','date','renderedContent'])
```

```python
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Text preprocessing functions from previous script
def tokenize(text):
    tokens = word_tokenize(text)
    return tokens

def lowercase(text):
    return text.lower()

def remove_unnecessary_char(text):
    text = re.sub('\n', ' ', text) # Remove every newline character
    text = re.sub(r'@[A-Za-z0-9_]+', ' ', text) # Remove every username
    text = re.sub('user', ' ', text) # Remove every occurrence of the word 'user'
    text = re.sub('rt', ' ', text) # Remove every retweet symbol
    text = re.sub('((www\.[^\s]+)|(https?://[^\s]+))', ' ', text) # Remove every URL
    text = re.sub('https?://t\.me/[^\s]+', ' ', text) # Remove every Telegram link
    text = re.sub ('[^0-9a-zA-Z]+', ' ', text) # Remove every non-alphanumeric character
    text = re.sub('  +', ' ', text) # Remove extra spaces
    return text

def normalize_alay(text):
    return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])

def normalize_alay2(text):
    return ' '.join([alay_dict_map2[word] if word in alay_dict_map2 else word for word in text.split(' ')])

def remove_stopword(text):
    text = ' '.join([ ' ' if word in id_stopword_dict.ada.values else word for word in text.split(' ')])
    text = re.sub('  +', ' ', text) # Remove extra spaces
    text = text.strip()
    return text

# Modified clean_text function that calls preprocessing functions from previous script
def preprocess(text):
    text = remove_unnecessary_char(text) # Remove unnecessary characters
    text = lowercase(text) # Convert text to lowercase
    text = normalize_alay(text) # Normalize text using alay_dict_map
    text = normalize_alay2(text) # Normalize text using alay_dict_map2
    text = remove_stopword(text) # Remove stopwords from text
    # tokens = tokenize(text)
    return text

# Apply the clean_text function to the 'Isi konten' column in datakorupsibaru
datakorupsibaru['renderedContent'] = datakorupsibaru['renderedContent'].apply(preprocess)

# Remove duplicates
datakorupsibaru.drop_duplicates(subset=['renderedContent'], inplace=True)

# Save the cleaned data to a new CSV file
datakorupsibaru.to_csv("cleaned_datakorupsibaru.csv", index=False)
```

```python
import pandas as pd
from textblob import TextBlob

# Baca dataset Twitter yang sudah di-preprocessing
twitter_df = pd.read_csv('result.csv')

# Fungsi untuk melakukan analisis sentimen menggunakan TextBlob
def analyze_sentiment(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity

    if sentiment > 0:
        return 'positive'
    elif sentiment < 0:
        return 'negative'
    else:
        return 'neutral'

# Analisis sentimen pada dataset Twitter
twitter_df['Sentiment'] = twitter_df['Tweet'].apply(analyze_sentiment)

# Menyimpan hasil ke file CSV
twitter_df.to_csv('result_sentiment.csv', index=False)
```

```python
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Baca dataset Twitter
twitter_df = pd.read_csv('result.csv')

# Baca dataset analisis positif dan negatif
abusive_df = pd.read_csv('abusive.csv')

# Inisialisasi stemmer Sastrawi
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan stemming pada teks
def stem_text(text):
    return ' '.join([stemmer.stem(word) for word in text.split()])

# Preprocessing pada dataset Twitter
twitter_df['Tweet'] = twitter_df['Tweet'].fillna('')  # Mengganti nilai null dengan string kosong
twitter_df['Tweet'] = twitter_df['Tweet'].apply(lambda x: str(x).lower())  # Ubah teks ke lowercase
twitter_df['Tweet'] = twitter_df['Tweet'].apply(stem_text)  # Lakukan stemming pada teks

# Preprocessing pada dataset analisis positif dan negatif
abusive_df['ABUSIVE'] = abusive_df['ABUSIVE'].apply(lambda x: str(x).lower())  # Ubah teks ke lowercase
abusive_df['ABUSIVE'] = abusive_df['ABUSIVE'].apply(stem_text)  # Lakukan stemming pada teks

# Menggabungkan dataset Twitter dengan dataset analisis positif dan negatif
twitter_df['Label'] = twitter_df['Tweet'].apply(lambda x: 'negatif' if x in abusive_df['ABUSIVE'].values else 'positif')

# Menyimpan hasil ke file CSV
twitter_df.to_csv('resultlabel.csv', index=False)
```
